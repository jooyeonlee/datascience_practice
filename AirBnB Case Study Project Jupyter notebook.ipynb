```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('files/AB_NYC_2019.csv')
air_bnb.head()
```

### How many neighborhood groups are available and which shows up the most?

```python
# 5 neighborhood groups and Manhattan shows up the most

air_bnb.describe()
air_bnb.groupby('neighbourhood_group').count()[['host_id']]
```

### Are private rooms the most popular in manhattan? 

```python
#Entire home/apt is the most popular in manhattan
print(air_bnb[(air_bnb.neighbourhood_group == 'Manhattan')].groupby('room_type').count()[['id']])
air_bnb[(air_bnb.neighbourhood_group == 'Manhattan')].groupby('room_type').sum('number_of_reviews')[['number_of_reviews']]
```

### Which hosts are the busiest and based on their reviews? 

```python
# Maya(host_id: 437312959)
print(air_bnb.groupby(air_bnb['host_id']).sum('number_of_reviews').sort_values('number_of_reviews', ascending=False)[['number_of_reviews']].head())
air_bnb[air_bnb['host_id']==37312959][['host_id', 'host_name', 'number_of_reviews']]
```

### Which neighorhood group has the highest average price?

```python
# Manhattan has the highest average price
air_bnb.groupby('neighbourhood_group').mean('price')[['price']].sort_values('price', ascending=False)
```

### Which neighbor hood group has the highest total price?

```python
def make_total(row):
    return row['price'] * row['minimum_nights']

air_bnb['total_price'] = air_bnb.apply(make_total, axis=1)
```

```python
# Manhattan
air_bnb.sort_values('total_price', ascending=False)[['host_name', 'neighbourhood_group', 'total_price']].head()
```

### Which top 5 hosts have the highest total price?

```python
# Individual hosting = Jenny, Katherine, Amy, Iveta, Noelle
# Consider all the host listing the host has, Blueground, Jenny, Kara, Kethrine, Amy
print(air_bnb.groupby(air_bnb['host_id']).sum('total_price').sort_values('total_price', ascending=False)[['total_price']].head())
air_bnb[air_bnb['host_id']==107434423][['host_id', 'host_name', 'price', 'minimum_nights', 'total_price', 'calculated_host_listings_count']]
air_bnb[air_bnb['host_id']==271248669][['host_id', 'host_name', 'price', 'minimum_nights', 'total_price', 'calculated_host_listings_count']]
air_bnb[air_bnb['host_id']==30283594][['host_id', 'host_name', 'price', 'minimum_nights', 'total_price', 'calculated_host_listings_count']]
air_bnb[air_bnb['host_id']==20582832][['host_id', 'host_name', 'price', 'minimum_nights', 'total_price', 'calculated_host_listings_count']]
air_bnb[air_bnb['host_id']==3906464][['host_id', 'host_name', 'price', 'minimum_nights', 'total_price', 'calculated_host_listings_count']]
```

### Who currently has no (zero) availability with a review count of 100 or more?

```python
# 162 host
air_bnb[(air_bnb['availability_365']==0) & (air_bnb['number_of_reviews'] >= 100)][['host_name', 'number_of_reviews', 'availability_365']].reset_index()
```

### What host has the highest total of prices and where are they located?

```python
# What host has the highest total of prices and where are they located?
#Individual hosting, Jenny has the highest total of prices and located in Manhattan
#Considering all of listing the host has, Blueground has the highest total of prices and located in Manhattan and Brooklyn
print(air_bnb.sort_values('total_price', ascending=False)[['host_name', 'neighbourhood_group', 'total_price']].head())
print(print(air_bnb.groupby(air_bnb['host_id']).sum('total_price').sort_values('total_price', ascending=False)[['total_price']].head()))
air_bnb[air_bnb['host_id']==107434423][['host_id', 'host_name', 'price', 'minimum_nights', 'total_price', 'calculated_host_listings_count', 'neighbourhood_group']]
```
