# Nba Regular Season 2018-19 Data Challenge

Your task will be to take the dataset given, and create an analysis answering the following 10 questions. This project will again test your knowledge of pandas in order to find the answers needed given the data you are presented with.

```python
import pandas as pd
import numpy as np

nba = pd.read_csv('./files/nbastats2018-2019.csv', sep=',')
nba
```

# What was the average age of player in the league?

```python
#25.9
nba['Age'].mean()
```

# What player scored the most points?

```python
# James Harden
nba.sort_values('Points', ascending=False)[['Name','Points']].head()
```

# What player had the most blocks during the season? Was it a post player (F/C)?

```python
# Myles Turner - Post Player (goal and rebound)
nba.sort_values('Blocks', ascending=False)[['Name', 'Points', 'Blocks', 'Rebounds']].head()
```

# Based on the regular season, who had the best chance to win a title given their win percentage?

```python
# James Harden
nba.sort_values(['WS','WS48'], ascending=False)[['Name', 'Points', 'WS', 'WS48']].head()
```

# What player had the best 3-pt percentage? 

```python
# Scott Machado in case of see the just FG3%
print(nba.sort_values(['FG3%', 'FGA'], ascending=False)[['Name', 'FG3%', 'FGA']].head())

# But when we consider the 3pt goal attempt number, James Harden is the bsst 3pt percentage
nba.sort_values(['FGA', 'FG3%'], ascending=False)[['Name', 'FG3%', 'FGA']].head()
```

# Who played the most minutes during the season

```python
#Bradley Beal
nba['MIN'] = nba['MP'] * nba['G']
nba.sort_values('MIN', ascending=False)[['Name','MP', 'G', 'MIN']].head()
```

# What player given their player effiecency rating was the clutchest during the season?

```python
#Giannis Antetokounmpo
# Remove outliers
nba[nba['MP']>=20].sort_values('PER', ascending=False)[['Name', 'PER', 'MP', 'Points', 'Blocks', 'Steals', 'Assists' ]].head()
```

# What team had the youngest roster?

```python
# Chicago Bulls
nba.groupby('Team').mean()[['Age']].sort_values('Age')
```

# Who is the highest paid player during the seasion?

```python
#print(type(nba['Salary'][1])) #Salary cannot sort becuase it is string type
nba['Salary'] = nba['Salary'].str.replace('-', '0') # replace '-' to 0
nba['Salary'] = nba['Salary'].astype(int) #'Type of Salary change into integer
```

```python
# Stephen Curry
nba.sort_values('Salary', ascending=False)[['Name', 'Salary']].head()
```
